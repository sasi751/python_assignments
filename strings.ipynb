```python
# Question 11: Find the most frequent character in "mississippi"
print("\nQuestion 11: Find the most frequent character in 'mississippi'")
s="mississippi"
d={}
for i in s:
  if i in d:
    d[i]+=1
  else:
    d[i]=1
ch=''
maxi=-10000007
for i,j in d.items():
  if j>maxi:
    maxi=j
    ch=i
print(ch)
```
