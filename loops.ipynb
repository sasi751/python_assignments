```python
s="madam"
if s==s[::-1]:
  print("palindrome")
else:
  print("nope")

ss="aaaaab"
i=0
j=len(ss)-1
while j>i:
  if ss[i]!=ss[j]:
    print("Nope")
    break
  i+=1
  j-=1
else:
  print("palindrome")
```

```python
l=[1,2,1,3,4,2,5,6,3]
l1=[]
for i in l:
  if l.count(i)>=2 and i not in l1:
    l1.append(i)
print(l1)

d={}
for i in l:
  if i not in d:
    d[i]=1
  else:
    d[i]+=1
print(d.items())
```
